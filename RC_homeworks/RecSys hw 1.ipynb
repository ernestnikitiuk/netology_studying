```python
import pandas as pd
```

```python
links = pd.read_csv("links.csv")
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
tags = pd.read_csv("tags.csv")
```

```python
links.head()
```

```python
movies.head()
```

```python
ratings.head()
```

```python
tags.head()
```

```python
movies_with_ratings = movies.join(ratings.set_index("movieId"), on = "movieId")
```

```python
movies_with_ratings.head()
```

```python
movies_with_ratings.groupby("title")[["rating"]].mean().sort_values("rating", ascending = False).head(20)
```

```python
movies_with_ratings.groupby("title")[["rating"]].sum().sort_values("rating", ascending = False).head(20)
```

```python
ratings_sum = movies_with_ratings.groupby("title")[["rating"]].sum()
```

```python
movies_with_tags = movies.join(tags.set_index("movieId"), on = "movieId")
```

```python
movies_with_tags.head()
```

```python
tags_count = movies_with_tags.groupby("title")[["tag"]].count()
```

```python
tags_count["tag"] = tags_count["tag"]+ 1
```

```python
results_df = ratings_sum.join(tags_count)
```

```python
results_df.head()
```

```python
results_df["res"] = results_df["rating"]*results_df["tag"]
```

```python
results_df.sort_values("res", ascending = False).head(20)
```

```python
results_df = results_df.join(movies.set_index("title"), on = "title")
results_df = results_df.sort_values("res", ascending = False)
```

```python
#Превращаем жанры в список причем уже в отсортированном датафрейме
```

```python
results_df["genres"] = results_df.apply(lambda x: x["genres"].split("|"), axis = 1)
```

```python
results_df = results_df.drop("movieId", axis = 1)
```

```python
#Выбираем только комедии
```

```python
comedy_results_df = results_df[results_df.apply(lambda x: pd.Series("Comedy").isin(x.genres), axis =1)[0]]
```
